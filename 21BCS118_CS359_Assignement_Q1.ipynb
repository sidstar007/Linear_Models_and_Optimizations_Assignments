```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("./MA201_marks.csv")
```

```python
marks_details = dataset.iloc[:, -5:]
```

```python
print(marks_details)
```

```python
x = marks_details.iloc[:, :-1].values
y = marks_details.iloc[:, -1].values
```

```python
print("Features: ", x)
print("Target: ", y)
```

```python
x_mean = [np.sum(i)/4 for i in x]
```

```python
marks_details['x_mean'] = x_mean
```

```python
dataset = marks_details
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(np.array(x_mean).reshape(69, 1), y.reshape(69, 1), test_size=0.2, random_state=42)
```

```python
x_mean = x_train
```

```python
print("X_Mean: ", x_mean)
```

```python
X = np.concatenate((np.ones((x_train.shape[0], 1)), x_train), axis=1)
```

```python
print(X, X.shape)
```

```python
w_cap = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y_train)
```

```python
print(w_cap)
```

```python
print(X.shape , w_cap.shape)
```

```python
y_pred = np.dot(X, w_cap)
```

```python
print("Y_pred: ", y_pred)
```

```python
y_real = y_train
```

```python
print(y_real , y_pred)
```

```python
training_loss_mse = np.mean((y_real - y_pred)**2)
print("Training Loss: ", training_loss_mse)
```

```python
X = np.concatenate((np.ones((x_test.shape[0], 1)), x_test), axis=1)
print(X, X.shape)
```

```python
w_cap = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y_test)
print(w_cap)
```

```python
y_pred = np.dot(X, w_cap)
print(y_pred)
```

```python
y_real = y_test
```

```python
testing_loss = np.mean((y_real - y_pred)**2)
print("Testing Loss:",testing_loss)
```
